```python
# 线性回归模型
import numpy as np
# 实际比赛数据
actual_scores = [
    (0, 2), (2, 1), (1, 2), (0, 0), (1, 0), (1, 5),
    (2, 0), (2, 0), (1, 2), (0, 1), (0, 1), (2, 1),
    (1, 0), (1, 4), (1, 5), (2, 2), (3, 0), (1, 2),
    (1, 3), (0, 2), (2, 3), (2, 0), (3, 1), (2, 2),
    (0, 2), (1, 6)
]

# 预测数据
predicted_scores = [
    (3.0, 1.0), (3.0, 0.0), (1.0, 1.0), (2.0, 2.0), (4.0, 1.0), (1.0, 1.0),
    (3.0, 1.0), (2.0, 0.0), (2.0, 1.0), (1.0, 1.0), (3.0, 1.0), (3.0, 1.0),
    (1.0, 2.0), (0.0, 2.0), (3.0, 1.0), (3.0, 2.0), (2.0, 1.0), (4.0, 1.0),
    (3.0, 1.0), (2.0, 1.0), (5.0, 1.0), (0.0, 1.0), (3.0, 1.0), (1.0, 2.0),
    (0.0, 2.0), (0.0, 2.0)
]

total_matches = len(actual_scores)

correct_score_predictions = 0
correct_outcome_predictions = 0

for actual, predicted in zip(actual_scores, predicted_scores):
    actual_home_goals, actual_away_goals = actual
    predicted_home_goals, predicted_away_goals = predicted
    
    # 检查整体预测准确率
    if actual_home_goals == predicted_home_goals and actual_away_goals == predicted_away_goals:
        correct_score_predictions += 1
    
    # 检查胜平负预测准确率
    actual_outcome = actual_home_goals - actual_away_goals
    predicted_outcome = predicted_home_goals - predicted_away_goals
    
    if (actual_outcome > 0 and predicted_outcome > 0) or \
       (actual_outcome == 0 and predicted_outcome == 0) or \
       (actual_outcome < 0 and predicted_outcome < 0):
        correct_outcome_predictions += 1

score_accuracy = correct_score_predictions / total_matches
outcome_accuracy = correct_outcome_predictions / total_matches

print("整体预测准确率: {:.2%}".format(score_accuracy))
print("胜平负预测准确率: {:.2%}".format(outcome_accuracy))
```
